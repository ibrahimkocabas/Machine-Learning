## K-Means (K-Ortalamalar)

**Amaç**, gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır. Kümeleme yöntemlerinin genel bir amacıdır. Benzerlik matrislerini kullanarak çeşitki uzaklık hesaplamaları yaparak gözlemler ya da değişkenler kümelenmeye çalışılmaktadır.Oluşturulmaya çalışan kümelerin kendi içinden homojen,birbirlerine göre ise heterojen olması istenir.

Kümeleme yöntemleri;hiyerarşik ve hiyerarşik olmayan kümeleme yöntemleri başlıkları altında incelenir.Bu açıdan bakıldığında **K-Means hiyerarşik olmayan** kümeleme yöntemidir.

**K-Means Adımları**
 * **1.Adım:** Küme sayısı belirlenir.
 * **2.Adım:** Rastegele k merkez seçilir.
 * **3.Adım:** Her gözlem için k merkezlere uzaklıklar hesaplanır ve gözlemler kendisine en yakın k merkezine atanır.
 * **4.Adım:** Her gözlem en yakın olduğu merkeze yani kümeye atanır.
 * **5.Adım:** Atama işlmelerinden sorna oluşan kümeler için tekrar merkez hesaplamaları yapılır.
 * **6.Adım:** Bu işlem belerilene bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamı(total within-cluster variation) minumum durumdaki gözlemlerin kümelenme yapısı nihai kümelenme yapısı olarak seçilir.

## Hiyerarşik Kümeleme

**Amaç;** gözlemleri birbirlerine olan benzerliklerine göre alt kümelere ayırmaktır.

K-Means'de ki amacımıza benzer bir şekilde bağımlı değişkeni olmayan gözlem birimlerini birbirlerine benzeterek bu benzetmeyi uzaklık metrikleri ile yaparak alt gruplara/kümelere ayırmak şeklinde gerçekleşmiş olacaktır.

**--** Hiyerarşik yöntemlerde küme sayısına **Dendrogram** sonuçlarına bakılarak karar verilir.

**--** Hiyerarşik olmayan yöntemlerde ise küme sayısı uygulama yapılmadan önce belirlenir.

**Hiyerarşik Kümeleme Yöntemleri'nde** veri seti gözlemler ya da değişkenler bazında kümeleme işlemine sokulabilirken, **Hiyerarşik olmayan kümeleme yöntemleri'nde** ise sadece gözlemlerin kümelenmesi mümkündür.

**Birleştirici Kümeleme Adımları**

Başlangıçta gözlem sayısı kadar küme vardır.

 * **1.Adım:** Veri setinde birbirine en yakın olan iki gözlem bulunur.
 * **2.Adım:** Bu iki nokta bir araya getirilerek yeni bör gözlem oluşturulur.Yani artık veri seti ilk birleşimdeki gözlemlerden oluşmaktadır.
 * **3.Adım:** Aynı işlem tekrarlanarak yukarı doğru çıkılır.Yani iki kümenin birleşiminden oluşan bu yeni kümeler aynı şekilde birbirlerine benzerliklerine göre tekrar birleştirilir.Bu işlem tüm gözlemler tek tek bir küme de toplanana kadar bu işlmeler tekrar edilir.

**Bölümleyici Kümeleme Adımları**

Başlangıçta 1 tane küme vardır, o da tüm veri setidir.

 * **1.Adım:** Tüm gözlemlerin bir arada olduğu küme iki alt kümeye ayrılır.
 * **2.Adım:** Oluşan yeni kümeler birbirlerine benzemeyen alt kümelere bölünür.
 * **3.Adım:** Aynı işlem gözlem sayısı kadar küme elde edilinceye kadar tekrar edilir.


**Hiyerarşik** vs **Hiyerarşik Olmayan** vs **Karar Ağaçları**

**--** Hiyerarşik yöntemlerde küme sayısına dendrogram sonuçlarına bakılarak karar verilirken,hiyerarşik olmayan yöntemlerde küme sayısı uygulama yapılmadan önce belirlenir.

**--** Hiyerarşik kümeleme yöntemlerinde veri seti gözlemler ya da değikenler bazında kümeleme işlemine sokulabilirken,hiyerarşik olmayan yöntemlerde sadecde gözlemlerin kümelenmesi mümkündür.

**--** Karar ağaçlarından farkı;karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken burada bağımlı değişken olmadığında dolayı gözlemler bağımsız değişkenkler üzerinden uzaklık hesaplarına göre kümelere ayrılır.

### Uygulama 

```python
df = pd.read_csv("USArrests.csv")
df.copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
```

**Amacımız eyaletleri belirli suç istatistiklerine ve nüfusla ilgili bir değişkene karşılık belirli bir kümelere ayırmak olacaktır.**

```python
from scipy.cluster.hierarchy import linkage  # istatistik kütüphanesi-kümeleme için
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
hc_single = linkage(df, "single")
```

```python
from scipy.cluster.hierarchy import dendrogram  # kümeleme sayısına karar vermek için

plt.figure(figsize=(15, 10))
plt.title('Hiyerarşik Kümeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklık')
dendrogram(
    hc_complete,
    leaf_font_size=10
);
```

```python
# İstenilen küme sayısında Dendrogram'mı budama/kesme
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15, 10))
plt.title('Hiyerarşik Kümeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklık')
dendrogram(
    hc_complete,
    truncate_mode = "lastp",  # istenilen değer argümanı
    p = 4,  # istenilen kümeleme sayısı
    show_contracted = True
);
```

Yukarıda ki dendrogram da istediğimiz gibi 4 küme sayısına ait görsel bulunmaktadır.

Ve bu kümelerin altında kaçar tane eleman olduğu bilgisi vermektedir.

### Optimum Küme Sayısı

```python
from scipy.cluster.hierarchy import dendrogram  # kümeleme sayısına karar vermek için

plt.figure(figsize=(15, 10))
plt.title('Hiyerarşik Kümeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklık')
dendrogram(
    hc_complete,
    leaf_font_size=10
);
```

```python
# Birleştirici Kümeleme İşlemi
from sklearn.cluster import AgglomerativeClustering  # birleştirici kümeleme
cluster = AgglomerativeClustering(n_clusters = 4, 
                                  affinity = "euclidean", # merkeze olan uzaklık/yakınlık hesaplama türü
                                  linkage = "ward")  # metod

cluster.fit_predict(df)
```

```python
# Gözlemlerin hangi küme sınıfı içinde oldukları 
pd.DataFrame({"Eyaletler": df.index, "Küme": cluster.fit_predict(df)})[0:10]
```

```python
# Veri seti içerisine küme numaralarını ekleme
df["cluster_number"] = cluster.fit_predict(df)
```

## Temel Bileşen Analizi (PCA)

**Temel fikir**, çok değişkenkli verinin ana özelliklerini daha az sayıda değişken/bileşen ile temsil etmektir.Diğer bir ifade ile;küçük miktarda bir bilgi kaybını göze alıp değişken boyutunu azaltmaktır.

Oluşan bileşenler arasında korelasyon yoktur.

Değişken gruplarının varyansslarını ifade eden özdğerler ile veri setindeki değişkenkler gruplandırılır.Gruplar arasında en fazla varyansa sahip gruplar en önemli gruplardır ki bunlar asal bileşenlerdir.

### Uygulama

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
df.head()
```

Temel bileşen analizini uygulayabilmek adına ilk olarak standartlaştırma işlemi yapmak gerekmektedir.Standardizasyon ise veriyi aynı ortalama (0) ve aynı standart sapmaya (1) sahip olması için yeniden ölçeklendirilmesidir.

```python
from sklearn.preprocessing import StandardScaler
df = StandardScaler().fit_transform(df)
df[0:5,0:5]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_fit = pca.fit_transform(df)
```

```python
# İsimlendirme
bilesen_df = pd.DataFrame(data = pca_fit,
                          columns = ["birinci_bilesen","ikinci_bileşen"])
```

```python
bilesen_df.head()
```

**Yaptığımız işlemin amacı;** en başta veri setimizde bulunan 4 değişkenin ifade ettiği bilgileri PCA ile standartlaştırarak ve indirgeyerek iki bileşen üzerinden temsil ederek aynı yapıyı kurmuş olduk.

-- Şimdi bu değişkenlerin yani indirgeme işlemi yaparken kaç değişken sayısı olacağına nasıl karar vereceğiz konusu ayrı bir konu. Mesela örneğimizde biz iki olarak yaptık.Gerçek veri setinin taşıdığı bilginin ne kadarını taşıdığı veya temsil ettiği bilgisin erişmek istersek;

```python
pca.explained_variance_ratio_
```

**Genelde çalışmalarda iki veya üç bileşen çok yüksek ihtimalle mecvut değişkenliği açıklamak için yeterli olacaktır.**

Bileşen sayısını bilmediğimizi ve buna nasıl karar vereceğimizi düşünürsek;

İlk olarak bileşen sayısını belirtmeden bir model nesnesi oluşturacağız ve daha sonra plot yardımı ile kaç bileşenin bize veri setini açıklayabilmede yardımcı olacaktır.

```python
pca = PCA().fit(df)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_));
```
